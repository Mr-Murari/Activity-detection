**EXTRACTING TEXT FROM THE URL**

```python
from newspaper import Article
url=pd.read_csv("/content/Input.csv")
directory = "C:\\Users\\MURARI\\Desktop\\BLACK\\"
for i in zip(url['URL_ID'],url['URL']) :
  
      # download and parse article
      article = Article(i[1])
      article.download()
      article.parse()
      #print(i)
      text=article.text
      
      filepath = directory + str(i[0])
      with open(filepath, 'w') as f:
         f.write(text)
```

**COMBING ALL STOP WORD THAT NEED TO BE EXCLUDED**

```python

directory = '/content/sample_data/StopWords'
stop_contents=[]
total_stop_bag=[]
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)

    if os.path.isfile(file_path):
        with open(file_path, 'r',encoding="ISO-8859-1") as file:
            contents = file.read()
            contents=contents.split()
            stop_contents.append(contents)
for stop in stop_contents:
  for ele in stop:
    total_stop_bag.append(ele)
print(total_stop_bag)
```

**VARIABLES CALCULATION**

```python
location='/content/C:\\Users\\MURARI\\Desktop\\BLACK\\'
positive_score_list=[]
negative_score_list=[]
polarity_score_list=[]
total_bag=[]
word_bag=[]
before_analy=[]
count_a_clean=[]
for id in (url['URL_ID']):
  data=location+str(id)
  with open(data, "r") as word_list:
      words = list(word_list.read().split())
      #word_bag.append(words)
      positive_score=int(len(set(words)&set(positive_list)))
      negative_score=len(set(words)&set(negative_list))
      polarity_score=(positive_score-negative_score)/((positive_score+negative_score)+0.000001)
      
      positive_score_list.append(positive_score)
      negative_score_list.append(negative_score)
      polarity_score_list.append(polarity_score)
```

**APPENDING VARIABLES INTO CSV FILES/EXCEL FILE**

```python
data_new=pd.read_csv('/content/Final_output.csv')
data_new['POSITIVE_SCORE'],data_new['Negative_SCORE'],data_new['polarity_SCORE'],=positive_score_list,negative_score_list,polarity_score_list
data_new.to_csv('/content/Final_output.csv')
```
